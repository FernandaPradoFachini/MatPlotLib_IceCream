```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data = {'Temperature': [37,40,49,61,72,79,83,81,75,64,53,42],
                         'Sales': [292,228,324,376,440,496,536,556,496,412,324,320]})
df
```

```python
df = df.sort_values('Temperature', ascending = False).reset_index(drop =True)
```

```python
#converter para celsius
celsius = (df['Temperature']-32)*5/9
df['Temperature'] = celsius
df
```

```python
df.to_csv('IceCream_Celsius.csv')#salvar alteracoes

plt.title('Ice Cream Truck')

plt.plot(df['Temperature'],df['Sales'],'r.--', label = 'US$')


plt.xticks(df['Temperature'])

plt.xlabel('Temperatura (Celsius)')
plt.ylabel('Sales (US$)')

plt.savefig('SelesIceCream.pdf') #salvar imagem no formato pdf

plt.legend()
plt.show()
```
